```python
import pandas as pd

# Load the JSON file
df = pd.read_json("Dataset for Detection of Cyber-Trolls.json", lines=True, orient='columns')

def extract_label(annotation):
        return int(annotation['label'][0])  
df['label'] = df['annotation'].apply(extract_label)

df.head(10000)
df.drop("annotation", inplace = True, axis =1)
df.drop("extras", inplace = True, axis =1)

df.to_csv('dataset.csv')
df
```
